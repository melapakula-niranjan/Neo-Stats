```python
import pandas as pd
import hashlib
```

```python
file_path = "/content/Contoso Data.xlsx"
```

```python
crm = pd.read_excel(file_path, sheet_name="crm_customers")
crm.head()
```

```python
transactions = pd.read_excel(file_path, sheet_name="core_transactions")
transactions.head()
```

```python
campaigns = pd.read_excel(file_path, sheet_name="marketing_campaigns")
campaigns.head()
```

```python
events = pd.read_excel(file_path, sheet_name="campaign_events")
events.head()
```

```python
crm["updated_at"] = pd.to_datetime(crm["updated_at"])
```

```python
last_watermark = pd.to_datetime("2024-01-01")
```

```python
crm_incremental = crm[crm["updated_at"] > last_watermark]
crm_incremental
```

```python
def generate_customer_key(val):
    return hashlib.sha256(val.encode()).hexdigest()

crm_incremental["CustomerKey"] = crm_incremental["customer_id"].apply(generate_customer_key)

crm_incremental[["customer_id", "CustomerKey"]].head()
```

```python
invalid_amount_txns = transactions[transactions["txn_amount"] < 0]
invalid_amount_txns
```

```python
missing_customer_txns = transactions[transactions["customer_id"].isnull()]
missing_customer_txns
```

```python
transactions["txn_timestamp"] = pd.to_datetime(transactions["txn_timestamp"])

future_txns = transactions[
    transactions["txn_timestamp"] > pd.Timestamp.now()
]

future_txns
```
